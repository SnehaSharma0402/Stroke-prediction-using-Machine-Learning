```python
import numpy as np
import pandas as pd
data=r"C:\Users\DELL\OneDrive\Documents\Documents\Project\project 2\healthcare-dataset-stroke-data.csv"
df=pd.read_csv(data)
df
```

```python
df.replace('N/A',np.nan,inplace=True)
df
```

```python
df.columns
```

```python
# Calculate frequency for each category in categorical columns
for col in ['work_type', 'Residence_type', 'smoking_status']:
    freq_encoding = df[col].value_counts() / len(df)
    df[col + '_freq'] = df[col].map(freq_encoding)

# Drop original categorical columns
df.drop(columns=['work_type', 'Residence_type', 'smoking_status'], inplace=True)
```

```python
df
```

```python
# Define columns to impute
numerical_cols = ['age', 'avg_glucose_level', 'bmi']
categorical_cols = ['gender','ever_married']

# Impute numerical columns with median and categorical columns with mode
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = imputer.fit_transform(df[categorical_cols])

df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])
df['ever_married'] = label_encoder.fit_transform(df['ever_married'])
df
```

```python
stroke = df['stroke']
df.drop(columns=['stroke'], inplace=True)
```

```python
df = pd.concat([df, stroke], axis=1)
```

```python
df
```

```python
df.to_csv('final.csv')
```

```python
id_column = df['id']
df.drop(columns=['id'], inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# Prepare the data
X = df.drop(columns='stroke')
y = df['stroke']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply SMOTE
smote = SMOTE(random_state=2)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Create DataFrame for the resampled data
resampled_data = pd.DataFrame(X_train_resampled, columns=X.columns)
resampled_data['stroke'] = y_train_resampled

# Save the resampled data to an Excel file
resampled_data.to_excel('resampled_smote.xlsx', index=False)

# Optionally, save the test data as well
# pd.DataFrame(X_test_scaled, columns=X.columns).to_excel('test_features.xlsx', index=False)
# pd.DataFrame(y_test).to_excel('test_target.xlsx', index=False)
```
