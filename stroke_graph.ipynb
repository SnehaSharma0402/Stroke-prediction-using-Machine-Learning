```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr = df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='stroke', data=df)
plt.title('Count of Stroke Cases')
plt.show()
```

```python
age_group = df.groupby('age')['stroke'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(age_group['age'], age_group['stroke'], marker='o', linestyle='-')
plt.title('Age-wise Stroke Prevalence')
plt.xlabel('Age')
plt.ylabel('Stroke Prevalence')
plt.grid(True)
plt.show()
```

```python
bins = [0, 30, 40, 50, 60, 70, 80, 90]
labels = ['0-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90']

# Create a new column for age bins
df['age_bin'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Group by age bins and calculate stroke prevalence
age_bin_group = df.groupby('age_bin')['stroke'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(age_bin_group['age_bin'], age_bin_group['stroke'], color='orange')
plt.title('Stroke Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Stroke Prevalence')
plt.grid(axis='y')
plt.show()
```

```python
# Define age bins and labels
bins = [0, 30, 40, 50, 60, 70, 80, 90]
labels = ['0-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90']

# Create a new column for age bins
df['age_bin'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Group by age bins and calculate stroke prevalence
age_bin_group = df.groupby('age_bin')['stroke'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(age_bin_group['age_bin'], age_bin_group['stroke'], color='orange')
plt.title('Stroke Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Stroke Prevalence')

# Remove borders
ax = plt.gca()  # Get current axis
for spine in ax.spines.values():
    spine.set_visible(False)

# Set y-axis limit
plt.ylim(0, 0.9)

# Add grid for y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
age_bin_stroke = df.groupby('age_bin')['stroke'].value_counts().unstack().fillna(0)
age_bin_stroke.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'salmon'])
plt.title('Stroke Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.grid(axis='y')
plt.show()
```

```python
# Group by age_bin and stroke status, and compute counts
age_bin_stroke = df.groupby('age_bin')['stroke'].value_counts().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10, 6))
ax = age_bin_stroke.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])

# Customize plot
plt.title('Stroke Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.grid(axis='y')

# Set y-axis limit to 3000
ax.set_ylim(0, 3000)

# Remove plot borders
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
gender_stroke = df.groupby('gender')['stroke'].value_counts().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10, 6))
gender_stroke.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.title('Stroke Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.grid(axis='y')
plt.show()
```

```python
df['gender'] = df['gender'].map({1: 'Male', 0: 'Female'})

# Group by gender and stroke status
gender_stroke = df.groupby('gender')['stroke'].value_counts().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10, 6))
gender_stroke.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.title('Stroke Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['No Stroke', 'Stroke'])
plt.grid(axis='y')
plt.show()
```
